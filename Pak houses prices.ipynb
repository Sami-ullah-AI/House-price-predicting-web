```python
my_data = sami.drop(
    ['property_id','location_id','page_url','location','latitude','longitude','locality','area',
     'date_added','agency','agent','area_marla','area_sqft','year','price_bin','month','day','purpose'], axis = 1)
```

```python
my_data.head()
```

```python
my_data['property_type'] = my_data['property_type'].astype('category').cat.codes
my_data['city'] = my_data['city'].astype('category').cat.codes
my_data['province_name'] = my_data['province_name'].astype('category').cat.codes
```

```python
X = my_data.drop('price', axis=1)
y = my_data['price']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

# Training Model
model.fit(X_train,y_train)

# Model Summary
model.fit(X_train,y_train)
print("Score :",model.score(X_test,y_test))
```

```python
from sklearn import linear_model
br=linear_model.BayesianRidge()
br.fit(X_train,y_train)
print("Score :",br.score(X_test,y_test))
```

```python
# Model Building
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
lr = LinearRegression()

# Training Model
lr.fit(X_train,y_train)

# Model Summary
y_pred_lr = lr.predict(X_test)

r_squared = r2_score(y_test,y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_lr))
print("R_squared :",r_squared)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

# Training Model
rf.fit(X_train,y_train)

# Model Summary
y_pred_rf = rf.predict(X_test)

r_squared = r2_score(y_test,y_pred_rf)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_rf))
print("R_squared :",r_squared)
```

```python
from sklearn import *
from sklearn.ensemble import GradientBoostingRegressor
gbt = GradientBoostingRegressor()

# Training Model
gbt.fit(X_train,y_train)

# Model Summary
y_pred_gbt = gbt.predict(X_test)

r_squared = r2_score(y_test,y_pred_gbt)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_gbt))
print("R_squared :",r_squared)
```
